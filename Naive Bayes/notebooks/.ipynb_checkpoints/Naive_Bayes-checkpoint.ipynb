```python
# Load EDA Packages
import pandas as pd
import numpy as np
import re
```

```python
# Load Data Visualization Packages
import seaborn as sns
```

```python
# Load Data Preprocessing Packages
import neattext.functions as nfx
```

```python
# load ML Packages

#Estimators
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

#Transformers
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
# Load Dataset
df = pd.read_csv("data/emotion_dataset_2.csv")
```

```python
df.head(50)
```

```python
# Data Visualization
#Value Counts
df['Emotion'].value_counts()
```

```python
# Plotting
sns.countplot(x='Emotion',data=df)
```

```python
#Data Preprocessing
dir(nfx)
```

```python

# removing user handles
df['Clean_Text'] = df['Text'].apply(nfx.remove_userhandles)
# removing stopwords
df['Clean_Text'] = df['Text'].apply(nfx.remove_stopwords)
# removing special hashtags
df['Clean_Text'] = df['Text'].apply(nfx.remove_hashtags)
# removing special punctuations
#df['Clean_Text'] = df['Text'].apply(nfx.remove_punctuations)
df['Clean_Text'] = df['Text'].str.replace('[^\w\s]','')
```

```python
df.head(10)
```

```python
# replace uppercase to lowercase
df['Clean1_Text'] = df['Clean_Text'].str.lower()
df.head(10)
```

```python
#Load Machine Learning Pkgs
from sklearn.naive_bayes import MultinomialNB

#Vectorizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_confusion_matrix
```

```python
# Split Dataset

from sklearn.model_selection import train_test_split
```

### Build Features From Our Text

```python
Xfeatures = df['Clean1_Text']
ylabels = df['Emotion']
```

```python
Xfeatures
```

```python
# Vectorizer
cv=CountVectorizer()
X=cv.fit_transform(Xfeatures)
```

```python
# Get Features By Name
cv.get_feature_names()
```

```python
#Dense Array Numpy
X.toarray()
```

```python
# Split Dataset

X_train,X_test,y_train,y_test = train_test_split(X,ylabels,test_size=0.3,random_state=42)
```

### Building Naive Bayes Model

```python
nv_model = MultinomialNB()
nv_model.fit(X_train,y_train)
```

```python
# Accuracy of Trained Naive Bayes Model (method 1)

nv_model.score(X_train,y_train)
```

```python
# Accuracy of Test Naive Bayes Model (method 1)

nv_model.score(X_test,y_test)
```

```python
# Predictions
y_pred_for_nv = nv_model.predict(X_test)
```

```python
y_pred_for_nv
```

# Making a prediction
+ Vectorized the Text
+ Applied Model

```python
sample_text = ["Today was a nice day"]
```

```python
vect = cv.transform(sample_text).toarray()
```

```python
nv_model.predict(vect)
```

```python
# check for probability percentages/confidence score
nv_model.predict_proba(vect)
#Getting all classes to be displayed
nv_model.classes_


```

```python
np.max(nv_model.predict_proba(vect))
```

```python
def predict_emotion(sample_text,model):
    myvect = cv.transform(sample_text).toarray()
    prediction = model.predict(myvect)
    pred_proba = model.predict_proba(myvect)
    pred_percentage_for_all = dict(zip(model.classes_,pred_proba[0]))
    print("Prediction: {}, Prediction Score:{}".format(prediction[0],np.max(pred_proba)))
    return pred_percentage_for_all
    
```

```python
predict_emotion(sample_text,nv_model)
```

### Model Evaluation 

```python
# Classification Report
#joy has the most amount of enteries in the database
print(classification_report(y_test,y_pred_for_nv))
```

```python
# Confusion Matrix
confusion_matrix(y_test,y_pred_for_nv)
```

```python
# plotting confusion matrix
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plot_confusion_matrix(nv_model,X_test,y_test)
```
